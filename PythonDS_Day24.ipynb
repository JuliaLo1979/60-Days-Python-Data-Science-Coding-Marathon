```python
import os
import mpl_toolkits.basemap
import matplotlib.pyplot as plt
import numpy as np

#https://matplotlib.org/basemap/api/basemap_api.html
#projection的種類
print(mpl_toolkits.basemap.supported_projections)
```

```python
plt.figure(figsize=(4,4)) #figsize的單位為英寸

m=Basemap(projection='ortho',resolution = None, lat_0=23,lon_0=120)
'''Default image size is 5400x2700, which can be quite slow and use quite a bit of memory. 
The scale keyword can be used to downsample the image (scale=0.5 downsamples to 2700x1350)'''
m.bluemarble(scale=0.25) 
plt.show()
```

```python
m1 = Basemap(projection ='lcc', width=12000000, height=9000000, lat_1=45.,lat_2=55,lat_0=50,lon_0=-107, resolution = None)
m1.bluemarble(scale=0.25)
plt.show()
```

```python
m2= Basemap(projection = 'mill', llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180, resolution ='c') # 成圖設定resolution 

#創建邊線
m2.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m2.drawcountries(linewidth=2)
#畫出州界
m2.drawstates(color= 'b')
#畫出城市 --未安裝counties圖
#m2.drawcounties(color= 'r')

plt.title('Basemap Tutorial')
plt.show()
```
