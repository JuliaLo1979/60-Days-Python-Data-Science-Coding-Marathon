```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns

boys =[164, 176, 169, 169, 165, 175, 159, 151, 144, 160, 183, 165, 156, 170, 164, 173, 165, 163, 177, 171]
girls = [169, 170, 162, 154, 183, 173, 169, 167, 170, 185, 162, 175, 168, 151, 181, 170, 182, 156, 159, 160]

'''
計算集中(平均數, 中位數, 眾數), 離散(全距, 變異數, 百分位數), 分布(偏度, 峰度)
'''
# 計算統計量_平均數
mean_boy= np.mean(boys)
mean_girl= np.mean(girls)
print("男孩身高平均=",mean_boy)
print("女孩身高平均=",mean_girl)
print()

# 計算統計量_中位數
np_median_boy=np.median(boys,axis=None)
np_median_girls=np.median(girls,axis=None)
print("np_median_boy=",np_median_boy)
print("np_median_girls=",np_median_girls)
print()

# 統計量_眾數(如果有多個眾數，取最小的值當眾數)
mode_boy=stats.mode(boys,axis=None)
mode_girls=stats.mode(girls,axis=None)
print("男孩身高眾數=",mode_boy, end = '')
print("男孩身高眾數=",mode_boy[0][0])
print()

print("女孩身高眾數=",mode_girls,end = '')
print("女孩身高眾數=",mode_girls[0][0])
print()

#全距
#rangeV=max(boys)-min(boys)
def rangeV(x): 
  return(max(x)-min(x))
    
print('男孩身高 全距=',rangeV(boys))
print('女孩身高 全距=',rangeV(girls))
print()

# 變異數
print("男孩身高變異數=",statistics.variance(boys))
print("女孩身高變異數=",statistics.variance(girls))
print()

# python 百分位數
#np
print("男孩-90百分位數=",np.percentile(boys, 90))
print("男孩-50百分位數=",np.percentile(boys, 50))
print("男孩-20百分位數=",np.percentile(boys, 20))
print()

#stat
print("女孩-90百分位數=",stats.scoreatpercentile(girls, 90))
print("女孩-50百分位數=",stats.scoreatpercentile(girls, 50))
print("女孩-20百分位數=",stats.scoreatpercentile(girls, 20))
print()

#計算峰度和偏度
print("男孩身高峰度=",stats.skew(boys))
print("女孩身高峰度=",stats.skew(girls))
print()

print("男孩身高偏度=",stats.kurtosis(boys))
print("女孩身高偏度=",stats.kurtosis(girls))
print()

#最後，畫圖看分布
# 設定雙格畫板大小
plt.subplot(2, 1, 1)
plt.hist(boys,alpha=.4,bins=40)
plt.xlim([140, 190]) 
plt.axvline(x=mean_boy)
plt.title('boy,skewness={0},kurtosis={1}'.format(round(stats.skew(boys),2),round(stats.kurtosis(boys),2)))

# 設定雙格畫板大小
plt.subplot(2, 1, 2)
plt.hist(girls,alpha=.4,bins=40)
plt.xlim([140, 190]) 
plt.axvline(x=mean_girl)
plt.title('girl,skewness={0},kurtosis={1}'.format(round(stats.skew(girls),2),round(stats.kurtosis(girls),2)))

plt.subplots_adjust(wspace=0.5, 
                    hspace=0.5)

plt.show()
```
